```python
import pandas as pd

def check_invalid_values(data):
    # Check for invalid values in 'age' and 'tumor-size'
    invalid_age_rows = data[~data['age'].str.match(r'\d+-\d+')]
    invalid_tumor_size_rows = data[~data['tumor-size'].str.match(r'\d+-\d+')]

    # Check for value counts in other categorical columns
    invalid_menopause_values = data[~data['menopause'].isin(data['menopause'].unique())]
    invalid_node_caps_values = data[~data['node-caps'].isin(['yes', 'no'])]
    invalid_breast_values = data[~data['breast'].isin(['left', 'right'])]
    invalid_breast_quad_values = data[~data['breast-quad'].isin(['left_low', 'left_up', 'right_low', 'right_up', 'central'])]
    invalid_irradiat_values = data[~data['irradiat'].isin(['yes', 'no'])]

    # Concatenate all invalid rows
    invalid_rows = pd.concat([invalid_age_rows, invalid_tumor_size_rows, invalid_menopause_values, 
                              invalid_node_caps_values, invalid_breast_values, invalid_breast_quad_values,
                              invalid_irradiat_values])
    
    return invalid_rows

print("Rows with invalid Values:")
print(check_invalid_values(data))
```

```python
# Calculate number of rows that contain at least one '?' across all columns - help from chatGPT
data[data.eq('?').any(axis=1)].shape[0]
```

**<i>The above two code blocks were generated from chatGPT. I was trying to find an effective way to identify possible, invalid values present in each rows' columns, so I didn't have to risk constantly coming across different types of invalid values, once I had addressed one. From the output, we see that the only invalid value is '?' for some columns.**<i>

```python
import numpy as np

# We replace any present '?' with NaN
data.replace('?', np.nan, inplace=True)
```

```python
print("Rows with invalid Values:")
print(check_invalid_values(data))
```

```python
# Replace NaN categorical vals with mode
categorical_cols = ['age', 'tumor-size', 'inv-nodes', 'menopause', 'node-caps', 'breast', 'breast-quad', 'irradiat']
for col in categorical_cols:
    data[col] = data[col].fillna(data[col].mode()[0])
```

**<i>From blocks [13] and [14], we see that the only numerical column 'deg-malig' doesn't contain any invalid/NaN values; thus, we just address the NaN values present in the respective categorical column(s) by replacing them with the column's mode.<i>**

### Visualize dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in data.keys():
    plt.figure()  
    sns.boxplot(x=data[col]) 
    plt.xlabel(col)  
    plt.title(f'Boxplot of {col}') 
    plt.show() 
```

**<i>Majority of the boxplots don't show meaningful information about the distribution of data. One reason is because there are multiple columns that only contain two possible values (they are binary categorical variables), hence their boxplots being incomprehensive. Another obvious reason is that boxplots are primarily used to represent numerical data; however, for demonstration, we went ahead and generated these plots. As such, we will visualize more suitable plots like count plots and bar plots.<i>** 

**<i>However, for 'inv-nodes', we see that there are a number of outliers, and that most rows have 0-2 inv-nodes. It's important to note that although we are treating inv-nodes as a categorical variable, we use a boxplot to find apparent outliers - this will allow us later when we are creating models, such as the K-Nearest Neightbor, to identify whether the column data is "noisy" and thus choose our hyperparameter(s) appropriately.<i>**

### Perform one-hot encoding on categorical variables

```python
data[categorical_cols] = data[categorical_cols].astype("category")
data = pd.get_dummies(data, columns=categorical_cols, drop_first=True)
data.head()
```

## Part 2: Building and assessing models

```python
data.describe()
```

### Split dataset

```python
X = data.drop('class', axis=1)
y = data['class']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=1)
```

**<i>In determining the split size, we took into account the fact that our dataset only contains 217 rows, which is not many. Thus, we wanted to dedicate at least 70% to the training set. We tested 70/30, 75/25, and 80/20, and found that the 75/25 split was most optimal in our report metrics.<i>** 

```python
X_train.shape
```

```python
X_test.shape
```

### K-Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsClassifier

knn_search_accuracy = KNeighborsClassifier()
# Create a dictionary of all values we want to test for n_neighbors
# note the use of 'n_neighbors', which is the exact function parameter name to the
# KNeighborsClassifier constructor
# representing the hyperparameter we are looking to optimize
param_grid = {"n_neighbors": np.arange(1, 100)}
```

```python
from sklearn.model_selection import GridSearchCV
knn_gscv = GridSearchCV(knn_search_accuracy, param_grid, cv=5)
```

```python
# for each parameter in our param_grid, for each possible value,
# fit model to data using cv folds and compute the accuracy
knn_gscv.fit(X_train, y_train)

# After the model has been fit, check top performing n_neighbors value
knn_gscv.best_params_
```

```python
from sklearn.metrics import classification_report

best_knn = knn_gscv.best_estimator_

# Generate classification report for the test data
print(f"Performance on TEST\n*******************\n{classification_report(y_test, best_knn.predict(X_test))}")

# Generate classification report for the train data
print(f"Performance on TRAIN\n********************\n{classification_report(y_train, best_knn.predict(X_train))}")
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=1)
dt_model = DecisionTreeClassifier(random_state=1).fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
print(f"Performance on TEST\n*******************\n{classification_report(y_test, dt_model.predict(X_test))}")
print(f"Performance on TRAIN\n********************\n{classification_report(y_train, dt_model.predict(X_train))}")
```

### Naive Bayes

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=1)

from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# fit the LG model -- random_state is used to shuffle the data; max_iter is max # of iterations for solver to converge (default is 100)
nb_model = LogisticRegression(random_state=1, max_iter=1000).fit(X_train, y_train)
# print the report
print(f"Performance on TEST\n*******************\n{classification_report(y_test, nb_model.predict(X_test))}")
print(f"Performance on TRAIN\n********************\n{classification_report(y_train, nb_model.predict(X_train))}")
```
